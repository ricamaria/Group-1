```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import json
import os
```

```python
# Read master postcodes list and postcodes lat/long csv
path = "Resources/australian_postcodes.csv"
path2 = "Output/postcodes.csv"

df = pd.read_csv(path)
pc_df = pd.read_csv(path2)
```

```python
# Join CSVs
postcode_df = pc_df.join(df, on= "Postcode", how="left")
postcode_df
```

```python
# Select and filter required data
postcode_df = postcode_df[["Postcode","state","lat","long","RA_2016"]]

# Rename columns
postcode_df = postcode_df.rename(columns= {"state": "State",
                            "lat": "Latitude",
                            "long": "Longitude",
                            "RA_2016": "Remoteness Area"})

postcode_df
```

```python
postcode_df.to_csv("Output/postcode_df.csv")
```
